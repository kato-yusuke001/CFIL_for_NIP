```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
root_path = "../CFIL_for_NIP/train_data/20250312_demo/"
task_name = "persam_f_mask_image_only"
folder_path = os.path.join(root_path, task_name, "test/20250312_8000")

output_path = os.path.join(folder_path, "eval")
if not os.path.exists(output_path):
    os.makedirs(output_path)

files = os.listdir(folder_path)
N = 30

initial_image = cv2.imread(os.path.join(root_path, "initial_image.jpg"))
initial_image = cv2.cvtColor(initial_image, cv2.COLOR_BGR2RGB)
initial_image = cv2.resize(initial_image, None, fx=0.5, fy=0.5)
initial_image = cv2.line(initial_image, [0,0], [0,initial_image.shape[0]], (255,0,0),  100)
initial_image = cv2.line(initial_image, [0,0], [initial_image.shape[1], 0], (255,0,0),  100)
initial_image = cv2.line(initial_image, [0,initial_image.shape[0]], [initial_image.shape[1], initial_image.shape[0]], (255,0,0),  100)
initial_image = cv2.line(initial_image, [initial_image.shape[1], 0], [initial_image.shape[1], initial_image.shape[0]], (255,0,0),  100)

for i in range(N):
    file = files[i]
    input_image_path = os.path.join(folder_path, f"{i:04d}_input_image.jpg")
    output_image_path = os.path.join(folder_path, f"{i:04d}_output_image.jpg")
    if not os.path.exists(input_image_path):
        break

    input_image = cv2.imread(input_image_path)
    input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
    input_image = cv2.resize(input_image, None, fx=0.5, fy=0.5)
    
    output_image = cv2.imread(output_image_path)
    output_image = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)
    output_image = cv2.resize(output_image, None, fx=0.5, fy=0.5)

    fig = plt.figure(figsize=(6,4), dpi=300)

    fig.add_subplot(1,3,1).set_title("input")
    plt.imshow(input_image)
    plt.axis('off')

    fig.add_subplot(1,3,2).set_title("output")
    plt.imshow(output_image)
    plt.axis('off')

    fig.add_subplot(1,3,3).set_title("bottleneck")
    plt.imshow(initial_image)
    plt.axis('off')

    # fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    plt.savefig(os.path.join(output_path, f"{i:04d}_eval_image.jpg"), bbox_inches="tight")
    plt.show()
     
```
