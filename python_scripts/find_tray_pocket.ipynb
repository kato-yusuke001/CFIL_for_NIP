```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import numpy as np

origin1 = np.array([-0.2965,  -0.28325])
origin2 = np.array([ 0.05937, -0.28325])

digit = np.array([0.0603, 0.0477])

frame = np.array([10, 5])

coordinate1 = []
coordinate2 = []

for i in range(50): 
    #現在検査位置[1] = 初期検査位置[1] + 移動単位[1] * ROUNDDOWN((ポケット番号_現在値 - 1) / 枠配置[1])
    # 10個ごとにx座標を変更
    x1 = origin1[0] + digit[0] * int(i / frame[0]) 
    x2 = origin2[0] + digit[0] * int(i / frame[0])

    # IF(ROUNDDOWN((ポケット番号_現在値 - 1) / 枠配置[1]) % 2 == 0)
    #     現在検査位置[2] = 初期検査位置[2] - 移動単位[2] * ((ポケット番号_現在値 - 1) % 枠配置[1])
    # ELSE()
    #     現在検査位置[2] = 初期検査位置[2] - 移動単位[2] * (枠配置[1] - ((ポケット番号_現在値 - 1) % 枠配置[1]) - 1)
    # ENDIF()
    # 偶数行と奇数行で増加方向を反転 
    if(int(i / frame[0]) % 2 == 0):
        y1 = origin1[1] - digit[1] * int(i % frame[0])
        y2 = origin2[1] - digit[1] * int(i % frame[0])
    else:
        y1 = origin1[1] - digit[1] *(frame[0] - int(i % frame[0]) - 1)
        y2 = origin2[1] - digit[1] *(frame[0] - int(i % frame[0]) - 1)
    
    coordinate1.append([x1, y1])
    coordinate2.append([x2, y2])

coordinate1 = np.array(coordinate1)
coordinate2 = np.array(coordinate2)
print(coordinate2)

# 点p0に一番近い点を取得
def func_search_neighbourhood(p0, ps):
    L = np.array([])
    for i in range(ps.shape[0]):
        norm = np.sqrt( (ps[i][0] - p0[0])*(ps[i][0] - p0[0]) +
                        (ps[i][1] - p0[1])*(ps[i][1] - p0[1]) )
        L = np.append(L, norm)
    return np.argmin(L) ,ps[np.argmin(L)], np.min(L)


def main(x=0, y=0):
    # target coordinate
    _x = x
    _y = y
    target_coor = np.array([_x, _y])
    idx1, nearrest_coor1, minL1 = func_search_neighbourhood(target_coor, coordinate1)
    idx2, nearrest_coor2, minL2 = func_search_neighbourhood(target_coor, coordinate2)
    print( len( coordinate1), len(coordinate2) )
    print( idx1, nearrest_coor1[0], nearrest_coor1[1], minL1) # idx, (x, y, th)
    print( idx2, nearrest_coor2[0], nearrest_coor2[1], minL2) # idx, (x, y, th)
```

```python
main(0.27, -0.40)
```

```python
np.savez('coordinate.npz', tray1=coordinate1, tray2=coordinate2)
```
